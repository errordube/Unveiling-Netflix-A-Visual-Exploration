```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/netflix-titles.csv'
netflix_data = pd.read_csv(file_path)

```

```python

import seaborn as sns

# Count the number of titles by country
country_counts = netflix_data['country'].value_counts()

# Select the top 10 countries
top_10_countries = country_counts.head(10)

# Plotting a pie chart
plt.figure(figsize=(10, 8))
top_10_countries.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette("tab10"))
plt.title('Percentage of Netflix Content by Top 10 Producing Countries')
plt.ylabel('')  # Hide the y-label
plt.show()
```
