Next, we need to check the percentage of NA.

```python
for i in netflix_df.columns:
  na=netflix_df[i].isna().sum()
  percent=na/len(netflix_df)*100
  print("{} : {}".format(i,percent))
```

```python
for i in disney_df.columns:
  na=disney_df[i].isna().sum()
  percent=na/len(disney_df)*100
  print("{} : {}".format(i,percent))
```

```python
for i in prime_df.columns:
  na=prime_df[i].isna().sum()
  percent=na/len(prime_df)*100
  print("{} : {}".format(i,percent))
```
